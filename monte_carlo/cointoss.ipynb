```python
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt

n = 100 #number of simulations
k = 10 #number of flips

results = np.zeros([k,n])

for j in range(n):
    head_prob = 0.5
    for i in range(k):
        results[i,j] = np.random.binomial(1,head_prob)
        if results[i,j] == 1:
            head_prob += 0.05
        elif results[i,j] == 0:
            head_prob -= 0.05

# print(results)

number_of_heads = np.sum(results, axis=0)
# print(number_of_heads)

[y,x] = np.histogram(number_of_heads, np.arange(0,k+2,1))
x = x[:-1]

fig = plt.figure(num=None, figsize=(10,4), dpi = 100)

ax = fig.add_subplot(1,1,1)
plt.bar(x,y/n)
plt.xlabel("number of heads")
plt.ylabel("something else")
plt.title("Number of heads when flipping coin " +str(k)+" times and repeating simulation"+str(n)+"times")
ticksx = np.arange(0,k+1,1)
ax.set_xticks(ticksx)
plt.show
```
