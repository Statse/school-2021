```python
# Use matplotlib.pyplot
# Draw curve of a function, e.g., sinusoidal
# Name x and y labels
# Set title and legend
# Set grid lines on

import matplotlib.pyplot  as plt
import numpy as np

x = np.array(range(0,100))/10
y = np.sin(x)

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Our first figure')
plt.legend(['Sinusoidal'])
plt.grid(True)
plt.show()
```
