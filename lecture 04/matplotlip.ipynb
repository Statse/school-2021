```python
# Use matplotlib.pyplot
# Draw curve of a function, e.g., sinusoidal
# Name x and y labels
# Set title and legend
# Set grid lines on

import matplotlib.pyplot  as plt
import numpy as np

x = np.array(range(0,100))/10
y = np.sin(x)

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Our first figure')
plt.legend(['Sinusoidal'])
plt.grid(True)
plt.show()
```

```python
# Load London crime count data
# Calculate annual total number of crimes
# Plot histogram
# Plot curve
# How axis are by default? Is it ok?
# Set axis on a proper way

import pandas as pd
data = pd.read_csv("../data/london_crimes.csv")

crime_count = data.groupby(['year'])['number_of_cases'].sum()

plt.figure(num=None,  figsize=(12,4), dpi=100, facecolor='w', edgecolor='k')
plt.bar(crime_count.index, crime_count.values)
plt.show()

plt.figure(num=None,  figsize=(12,4), dpi=100, facecolor='w', edgecolor='k')
plt.plot(crime_count.index, crime_count.values)
plt.ylim([0,800000])
plt.show()
```
