```python
from tabulate import tabulate

longtermcostpergb = 0.01 #over 90 days old data
shorttermcostpergb = 0.02 #over 90 days old data
year = 365 
monthsperyear = 12
avgmonth = year/monthsperyear #avg days per month
dataperday = 10 #data per day
avgdatapermonth = avgmonth*dataperday #monthly data

#calculate how much data we have
longtermstorage = 0
shortermstorage = 0
totalstorage = 0

#calculate costs
longtermcost = 0
shorttermcost = 0
totalcost = 0
yearlycost = 0

yearstocalc = 5

for a in range(yearstocalc): # loop years
    print("========================================")
    print("================ YEAR:",a+1,"===============")
    print("========================================")
    for x in range(monthsperyear):  #loop script for every month
        currentmonth = x + 1 # since 1st index is 0
    
        totalstorage += avgdatapermonth
        shortermstorage += totalstorage

        if (shortermstorage > avgdatapermonth*3):# short term storage only contains data from 90 days
            shortermstorage = avgdatapermonth*3

        longtermstorage = totalstorage - shortermstorage # get how much data is in longterm storage

        if (longtermstorage < 0): # cant have negative storage
            longtermstorage = 0

        longtermcost += longtermstorage*longtermcostpergb #long term storage cost for this month
        shorttermcost += shortermstorage*shorttermcostpergb #short term storage cost for this month
        yearlycost = longtermcost + shorttermcost #yearly cost 
        totalcost = totalcost + longtermcost + shorttermcost #total cost 

        if (currentmonth == 12):
            print("COST FOR THE YEAR ", yearlycost)
            yearlycost = 0
            print(tabulate([
            ['Month', currentmonth, shortermstorage, shorttermcost, longtermstorage, longtermcost, totalstorage, totalcost]],
            headers=['Month', 'short-term storage (gb)','cost $', 'longterm storage(gb)', 'longterm cost $', 'total storage(gb)' ,'total cost $']
            ))

        # print(tabulate([
        #     ['Month', currentmonth, shortermstorage, shorttermcost, longtermstorage, longtermcost, totalstorage, totalcost]],
        #     headers=['Month', 'short-term storage (gb)','cost $', 'longterm storage(gb)', 'longterm cost $', 'total storage(gb)' ,'total cost $']
        #     ))
        # print("")
```
