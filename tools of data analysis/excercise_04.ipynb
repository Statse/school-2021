```python
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression

# data = px.data.gapminder().query("continent=='Europe'")
# data.head()
```

```python
# exercise 4
# How year, life expectation and GDP per capita are related? Draw scatter plots

finland = px.data.gapminder().query("country=='Finland'")
germany = px.data.gapminder().query("country=='Germany'")
turkey = px.data.gapminder().query("country=='Turkey'")

finland.head()

plt.scatter(finland["lifeExp"],finland["gdpPercap"], label = "FI")
plt.scatter(germany["lifeExp"],germany["gdpPercap"], label = "GER")
plt.scatter(turkey["lifeExp"],turkey["gdpPercap"], label = "TUR")

plt.xlabel("Life expectancy")
plt.ylabel("GDP")
plt.legend()
```

Here we can see pretty clear correlation between GDP and life expectancy. Both Finland and Germany have significantly higher GPD and life expectancy compared to Turkey.

```python
# Calculate linear regression of life expectation vs. time and life expectation vs. GDP in Europe, Germany and Turkey.
europe_data = px.data.gapminder().query("continent=='Europe'")

#time
x = europe_data["year"]
y = europe_data["lifeExp"]

x = x.values.reshape(-1,1)
y = y.values.reshape(-1,1)

reg = LinearRegression().fit(x, y)
y_reg = reg.predict(x)

fig = plt.figure(figsize = (12,9), dpi = 60)
plt.title("Year and life expectancy")
plt.xlabel("Year")
plt.ylabel("Life expectancy")
plt.grid(True)
plt.scatter(x,y)
plt.plot(x, y_reg)
plt.show()

#gdp
x = europe_data["lifeExp"]
y = europe_data["gdpPercap"]

x = x.values.reshape(-1,1)
y = y.values.reshape(-1,1)

reg = LinearRegression().fit(x, y)
y_reg = reg.predict(x)

fig = plt.figure(figsize = (12,9), dpi = 60)
plt.title("GDP and life expectancy")
plt.xlabel("Life expectancy")
plt.ylabel("GDP")
plt.grid(True)
plt.scatter(x,y)
plt.plot(x, y_reg)
plt.show()

```

With europe data it seems that linear regression is not the best way to analyze this. Age and GDP has exponential growth instead of linear growth.
