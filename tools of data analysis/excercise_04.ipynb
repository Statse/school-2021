```python
# exercise 4
# How year, life expectation and GDP per capita are related? Draw scatter plots

finland = px.data.gapminder().query("country=='Finland'")
germany = px.data.gapminder().query("country=='Germany'")
turkey = px.data.gapminder().query("country=='Turkey'")

finland.head()

plt.scatter(finland["lifeExp"],finland["gdpPercap"], label = "FI")
plt.scatter(germany["lifeExp"],germany["gdpPercap"], label = "GER")
plt.scatter(turkey["lifeExp"],turkey["gdpPercap"], label = "TUR")

plt.xlabel("Life expectancy")
plt.ylabel("GDP")
plt.legend()
```

Here we can see pretty clear correlation between GDP and life expectancy. Both Finland and Germany have significantly higher GPD and life expectancy compared to Turkey.
